```python
from transformers import pipeline
from transformers import AutoTokenizer
from optimum.onnxruntime import ORTModelForQuestionAnswering
from optimum.pipelines import pipeline
import sentencepiece

import numpy as np
```

```python
onnx_path = "/content/gdrive/MyDrive/Transformers_files/onnx_roberta_pipeline"
```

```python
model = ORTModelForQuestionAnswering.from_pretrained(onnx_path,file_name="model_quantized.onnx")
tokenizer = AutoTokenizer.from_pretrained(onnx_path)

modelo_pipe_roberta= pipeline("question-answering", model=model, tokenizer=tokenizer)
```

```python
def modelo_roberta_pipeline(pregunta, texto):

  respuesta_modelo = modelo_pipe_roberta(question = pregunta, context= texto)
  respuesta = respuesta_modelo['answer']
  score = str(np.round(respuesta_modelo['score'], 5))
  return respuesta + '  (score=' + score+ ')'
```
